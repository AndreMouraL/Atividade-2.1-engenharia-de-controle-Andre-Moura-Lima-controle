```python

# Parâmetros
Kp = 2
Ki = 100
Kd = 0.5

# Tempo
t = np.linspace(0, 5, 1000)

# Sinal de entrada: degrau unitário
def degrau(t):
    return np.ones_like(t)

e = degrau(t)

# Ação Proporcional
up = Kp * e

# Ação Integral
ui = Ki * np.cumsum(e) * (t[1] - t[0])  # Integração numérica

# Ação Derivativa
ud = Kd * np.gradient(e, t)

# Combinados
upi = up + ui
upd = up + ud
upid = up + ui + ud

# Plotagem
plt.figure(figsize=(14, 10))

plt.subplot(3, 2, 1)
plt.plot(t, up, label='P', color='blue')
plt.title('Ação Proporcional (P)')
plt.grid()

plt.subplot(3, 2, 2)
plt.plot(t, ui, label='I', color='orange')
plt.title('Ação Integral (I)')
plt.grid()

plt.subplot(3, 2, 3)
plt.plot(t, ud, label='D', color='green')
plt.title('Ação Derivativa (D)')
plt.grid()

plt.subplot(3, 2, 4)
plt.plot(t, upi, label='PI', color='purple')
plt.title('Ação Proporcional-Integral (PI)')
plt.grid()

plt.subplot(3, 2, 5)
plt.plot(t, upd, label='PD', color='brown')
plt.title('Ação Proporcional-Derivativa (PD)')
plt.grid()

plt.subplot(3, 2, 6)
plt.plot(t, upid, label='PID', color='red')
plt.title('Ação Proporcional-Integral-Derivativa (PID)')
plt.grid()

plt.tight_layout()
plt.show()
```

Se fosse derivativo: D

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do circuito
Rf = 2e3  # 2 kΩ
C = 0.1e-6  # 0.1 µF
Kd = Rf * C  # Ganho derivativo (0.0002 s)

# Tempo e sinal de entrada (degrau unitário)
t = np.linspace(-0.001, 0.01, 10000)  # Tempo de -1ms a 10ms (alta resolução)
Vi = np.where(t >= 0, 1, 0)  # Degrau unitário em t=0

# Versão teórica (impulso infinito em t=0)
delta_t = t[1] - t[0]  # Resolução temporal
impulso_teorico = np.zeros_like(t)
impulso_teorico[np.argmin(np.abs(t))] = 1 / delta_t  # Impulso em t=0

# Versão prática (filtrada para simular banda limitada)
tau = 1e-4  # Constante de tempo do filtro (simula limitação do amplificador)
derivada_pratica = np.gradient(Vi, t)  # Derivada numérica
derivada_filtrada = -np.exp(-t/tau) * (t >= 0)  # Resposta exponencial

# Saídas
Vo_teorico = -Kd * impulso_teorico  # Teórico (pico infinito)
Vo_pratico = -Kd * derivada_filtrada  # Prático (limitado)

# Plotagem
plt.figure(figsize=(10, 4))
plt.plot(t, Vi, '--', label='Entrada $V_i(t)$ (degrau)', color='gray')
plt.plot(t, Vo_teorico, label='Saída teórica (impulso)', color='red', linewidth=1, alpha=0.5)
plt.plot(t, Vo_pratico, label='Saída prática (limitada)', color='blue', linewidth=2)
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (V)')
plt.title('Resposta ao Degrau - Ação Derivativa Pura (D)')
plt.legend()
plt.grid(True)
plt.axvline(x=0, color='black', linestyle=':')
plt.ylim(-12, 12)  # Ajuste para visualizar o pico
plt.xlim(-0.001, 0.01)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do circuito
R1 = 1e3      # 1 kΩ
Rf = 2e3      # 2 kΩ
C = 1e-6      # 1 µF
Kp = 1 + Rf/R1  # Ganho proporcional (Kp = 3)
Ki = 1/(R1 * C) # Ganho integral (Ki = 1000 s⁻¹)

# Tempo e sinal de entrada (degrau unitário)
t = np.linspace(-0.1, 2, 1000)  # Tempo de -0.1s a 2s
Vi = np.where(t >= 0, 1, 0)      # Degrau unitário em t=0

# Componentes da saída
Vo_P = Kp * Vi                     # Parte proporcional (degrau em 3V)
Vo_I = Ki * t * (t >= 0)           # Parte integral (rampa)
Vo_PI = Vo_P + Vo_I                # Saída completa do PI

# Saturação prática (limite de ±15V)
Vo_PI_sat = np.clip(Vo_PI, -15, 15)

# Plotagem
plt.figure(figsize=(10, 5))
plt.plot(t, Vi, '--', label='Entrada $V_i(t)$ (degrau)', color='gray')
plt.plot(t, Vo_P, ':', label='Componente proporcional (P)', color='blue')
plt.plot(t, Vo_I, ':', label='Componente integral (I)', color='green')
plt.plot(t, Vo_PI, label='Saída PI (P + I)', color='red', linewidth=2)
plt.plot(t, Vo_PI_sat, '--', label='Saída PI saturada (±15V)', color='darkred')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (V)')
plt.title('Resposta ao Degrau - Controlador Proporcional-Integral (PI)')
plt.legend()
plt.grid(True)
plt.axvline(x=0, color='black', linestyle=':')
plt.ylim(-1, 16)
plt.xlim(-0.1, 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do circuito
R1 = 1e3      # 1 kΩ
Rf = 2e3      # 2 kΩ
C = 0.1e-6    # 0.1 µF
Kp = 1 + Rf/R1  # Ganho proporcional (Kp = 3)
Kd = Rf * C     # Ganho derivativo (Kd = 0.0002 s)

# Tempo e sinal de entrada (degrau unitário)
t = np.linspace(-0.001, 0.01, 10000)  # Alta resolução para capturar o pico
Vi = np.where(t >= 0, 1, 0)            # Degrau unitário em t=0

# Simulação do pico derivativo (prático, limitado pela banda do amplificador)
tau = 1e-4  # Constante de tempo do filtro (simula limitação física)
derivada = -np.exp(-t/tau) * (t >= 0)  # Resposta exponencial ao degrau

# Componentes da saída
Vo_P = Kp * Vi                     # Parte proporcional (degrau em 3V)
Vo_D = Kd * derivada               # Parte derivativa (pico em t=0)
Vo_PD = Vo_P + Vo_D                # Saída completa do PD

# Plotagem
plt.figure(figsize=(10, 5))
plt.plot(t, Vi, '--', label='Entrada $V_i(t)$ (degrau)', color='gray')
plt.plot(t, Vo_P, ':', label='Componente proporcional (P)', color='blue')
plt.plot(t, Vo_D, ':', label='Componente derivativa (D)', color='green')
plt.plot(t, Vo_PD, label='Saída PD (P + D)', color='red', linewidth=2)
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (V)')
plt.title('Resposta ao Degrau - Controlador Proporcional-Derivativo (PD)')
plt.legend()
plt.grid(True)
plt.axvline(x=0, color='black', linestyle=':')
plt.ylim(-1, 5)
plt.xlim(-0.001, 0.01)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do circuito PID
R1 = 1e3      # 1 kΩ (para ação P e I)
Rf = 2e3      # 2 kΩ (para ação P e D)
C1 = 0.1e-6   # 0.1 µF (ação derivativa)
C2 = 1e-6     # 1 µF (ação integral)

# Ganhos do PID
Kp = 1 + Rf/R1                # Ganho proporcional (3)
Ki = 1/(R1 * C2)              # Ganho integral (1000 s⁻¹)
Kd = Rf * C1                  # Ganho derivativo (0.0002 s)

# Tempo e sinal de entrada (degrau unitário)
t = np.linspace(-0.001, 0.1, 10000)  # Tempo de -1ms a 100ms
Vi = np.where(t >= 0, 1, 0)           # Degrau unitário em t=0

# Simulação da derivada (limitada por filtro passa-baixa)
tau = 1e-4  # Constante de tempo do filtro
derivada = -np.exp(-t/tau) * (t >= 0)  # Resposta exponencial

# Componentes da saída
Vo_P = Kp * Vi                     # Parte proporcional (degrau em 3V)
Vo_I = Ki * t * (t >= 0)           # Parte integral (rampa)
Vo_D = Kd * derivada               # Parte derivativa (pico)
Vo_PID = Vo_P + Vo_I + Vo_D        # Saída completa do PID

# Saturação prática (±15V)
Vo_PID_sat = np.clip(Vo_PID, -15, 15)

# Plotagem
plt.figure(figsize=(12, 6))
plt.plot(t, Vi, '--', label='Entrada $V_i(t)$ (degrau)', color='gray')
plt.plot(t, Vo_P, ':', label='Componente Proporcional (P)', color='blue')
plt.plot(t, Vo_I, ':', label='Componente Integral (I)', color='green')
plt.plot(t, Vo_D, ':', label='Componente Derivativa (D)', color='orange')
plt.plot(t, Vo_PID, label='Saída PID (P+I+D)', color='red', linewidth=2)
plt.plot(t, Vo_PID_sat, '--', label='Saída PID saturada (±15V)', color='darkred')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (V)')
plt.title('Resposta ao Degrau - Controlador PID (Proporcional-Integral-Derivativo)')
plt.legend(loc='upper right')
plt.grid(True)
plt.axvline(x=0, color='black', linestyle=':')
plt.ylim(-2, 20)
plt.xlim(-0.001, 0.1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Configurações gerais
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['font.size'] = 12

# Tempo de simulação
t = np.linspace(-0.1, 1, 5000)  # Intervalo ampliado para visualização
dt = t[1] - t[0]

# Sinal de entrada degrau unitário (com tratamento em t=0)
e_t = np.zeros_like(t)
e_t[t >= 0] = 1

# ---------------------------------------------------------------------------
# 1. CONTROLADOR PROPORCIONAL (P) - u(t) = Kp*e(t)
# ---------------------------------------------------------------------------
Kp = -10
u_p = Kp * e_t

plt.figure()
plt.plot(t, u_p, 'b-', linewidth=2, label=f'P: $u(t) = {Kp} \cdot u(t)$')
plt.title('Resposta do Controlador Proporcional (P) ao Degrau')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída $u(t)$')
plt.axhline(0, color='k', linestyle='--', alpha=0.3)
plt.axvline(0, color='k', linestyle='--', alpha=0.3)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.xlim([-0.1, 1])
plt.ylim([1.1*Kp, -0.5*Kp])
plt.show()

# ---------------------------------------------------------------------------
# 2. CONTROLADOR INTEGRAL (I) - u(t) = Ki*∫e(t)dt
# ---------------------------------------------------------------------------
Ki = -1000
integral = np.cumsum(e_t) * dt  # Integral numérica
u_i = Ki * integral

plt.figure()
plt.plot(t, u_i, 'r-', linewidth=2, label=f'I: $u(t) = {Ki} \cdot t$')
plt.title('Resposta do Controlador Integral (I) ao Degrau')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída $u(t)$')
plt.axhline(0, color='k', linestyle='--', alpha=0.3)
plt.axvline(0, color='k', linestyle='--', alpha=0.3)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.xlim([-0.1, 1])
plt.ylim([1.1*Ki*1, 0])  # Ajuste para rampa
plt.show()

# ---------------------------------------------------------------------------
# 3. CONTROLADOR DERIVATIVO (D) - u(t) = Kd*δ(t)
# ---------------------------------------------------------------------------
Kd = -0.01
# Criando um impulso (delta de Dirac) em t=0
impulse = np.zeros_like(t)
impulse[(t >= -dt/2) & (t <= dt/2)] = 1/dt  # Área unitária
u_d = Kd * impulse

plt.figure()
markerline, stemlines, baseline = plt.stem(t, u_d,
                                         label=f'D: $u(t) = {Kd} \cdot \delta(t)$',
                                         basefmt=' ')
plt.setp(markerline, color='g', markersize=10)
plt.setp(stemlines, color='g', linewidth=2)
plt.title('Resposta do Controlador Derivativo (D) ao Degrau')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída $u(t)$')
plt.axhline(0, color='k', linestyle='--', alpha=0.3)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.xlim([-0.1, 1])
plt.ylim([1.5*Kd/dt, -1.5*Kd/dt])
plt.show()

# ---------------------------------------------------------------------------
# 4. CONTROLADOR PI - u(t) = Kp*e(t) + Ki*∫e(t)dt
# ---------------------------------------------------------------------------
Kp = -10
Ti = 0.01  # Tempo integral
Ki = Kp/Ti
u_pi = Kp*e_t + Ki*integral

plt.figure()
plt.plot(t, u_pi, 'm-', linewidth=2,
        label=f'PI: $-10(1 + t/{Ti})u(t)$')
plt.title('Resposta do Controlador Proporcional-Integral (PI) ao Degrau')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída $u(t)$')
plt.axhline(0, color='k', linestyle='--', alpha=0.3)
plt.axvline(0, color='k', linestyle='--', alpha=0.3)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.xlim([-0.1, 1])
plt.ylim([1.2*np.min(u_pi), -0.2*np.min(u_pi)])
plt.show()

# ---------------------------------------------------------------------------
# 5. CONTROLADOR PD - u(t) = Kp*e(t) + Kd*δ(t)
# ---------------------------------------------------------------------------
Kp = -10
Td = 0.01  # Tempo derivativo
Kd = Kp*Td
u_pd = Kp*e_t + Kd*impulse

plt.figure()
plt.plot(t, Kp*e_t, 'orange', linestyle='--', linewidth=2, label='Componente P')
markerline, stemlines, baseline = plt.stem(t, Kd*impulse,
                                         label='Componente D',
                                         basefmt=' ')
plt.setp(markerline, color='crimson', markersize=10)
plt.setp(stemlines, color='crimson', linewidth=2)
plt.plot(t, u_pd, 'b-', linewidth=3, label='Saída PD')
plt.title('Resposta do Controlador Proporcional-Derivativo (PD) ao Degrau')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída $u(t)$')
plt.axhline(0, color='k', linestyle='--', alpha=0.3)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.xlim([-0.1, 1])
plt.ylim([1.1*Kp, -0.5*Kp])
plt.show()

# ---------------------------------------------------------------------------
# 6. CONTROLADOR PID - u(t) = Kp*e(t) + Ki*∫e(t)dt + Kd*δ(t)
# ---------------------------------------------------------------------------
u_pid = Kp*e_t + Ki*integral + Kd*impulse

plt.figure()
plt.plot(t, Kp*e_t, 'orange', linestyle='--', linewidth=2, label='Componente P')
plt.plot(t, Ki*integral, 'green', linestyle='--', linewidth=2, label='Componente I')
markerline, stemlines, baseline = plt.stem(t, Kd*impulse,
                                         label='Componente D',
                                         basefmt=' ')
plt.setp(markerline, color='purple', markersize=10)
plt.setp(stemlines, color='purple', linewidth=2)
plt.plot(t, u_pid, 'r-', linewidth=3, label='Saída PID')
plt.title('Resposta do Controlador PID ao Degrau')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída $u(t)$')
plt.axhline(0, color='k', linestyle='--', alpha=0.3)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.xlim([-0.1, 1])
plt.ylim([1.2*np.min(u_pid), -0.2*np.min(u_pid)])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control  # Certifique-se de ter a versão mais recente: pip install --upgrade control

# Parâmetros do circuito
R1 = 1000    # Resistência R1 (em Ohms)
R2 = 500     # Resistência R2 (em Ohms)
R3 = 2000    # Resistência R3 (em Ohms)
R4 = 3000    # Resistência R4 (em Ohms)
C1 = 1e-6    # Capacitância C1 (em Farads)
C2 = 2e-6    # Capacitância C2 (em Farads)

# Calculando os ganhos do controlador PID
Kp = (R4 * (R1 * C1 + R2 * C2)) / (R3 * R1 * C2)
Ki = R4 / (R3 * R1 * C2)
Kd = (R4 * R2 * C1) / R3

# Função de transferência do controlador PID
s = control.TransferFunction.s
G_pid = Kp + Ki / s + Kd * s

# Gerando a faixa de frequências
omega = np.logspace(-2, 3, 500)  # Frequências de 0.01 a 1000 rad/s

# Plotando o diagrama de Bode
plt.figure(figsize=(12, 8))
control.bode(G_pid, omega=omega, dB=True, Hz=False, deg=True)
plt.suptitle('Diagrama de Bode - Controlador PID', fontsize=16)
plt.tight_layout()
plt.show()

# Exibindo os valores dos ganhos
print(f"Ganho Proporcional (Kp): {Kp}")
print(f"Ganho Integral (Ki): {Ki}")
print(f"Ganho Derivativo (Kd): {Kd}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Parâmetros do circuito
R1 = 1e3     # 1 kΩ
R2 = 10e3    # 10 kΩ
R3 = 1e3     # 1 kΩ
R4 = 10e3    # 10 kΩ
C2 = 1e-6    # 1 µF

# Constantes
K = R4 / R3
Ti = R2 * C2

# Função de transferência: G(s) = K * (Ti*s + 1) / (R1*C2*s)
num = [K * Ti, K]
den = [R1 * C2, 0]

G = ctrl.TransferFunction(num, den)

# Resposta ao degrau
t, y = ctrl.step_response(G)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(t, y)
plt.title("Resposta ao Degrau do Controlador PI")
plt.xlabel("Tempo (s)")
plt.ylabel("Saída")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# --- Parâmetros do circuito PID analógico ---
R1 = 1.0
R2 = 1.0
R3 = 1.0
R4 = 1.0
C1 = 1.0
C2 = 1.0

# --- Cálculo dos parâmetros PID ---
Kp = R4 * (R1 + R2 * C2) / (R3 * R1 * C2)
Ti = 1 / (R1 * C1 + R2 * C2)
Td = (R1 * C1 * R2 * C2) / (R1 * C1 + R2 * C2)

# Filtro derivativo (opcional)
alpha = 0.01  # Ajustável para suavizar derivada

# --- Função de transferência do PID ---
s = ctrl.TransferFunction.s

# Termos proporcional, integral e derivativo com filtro
P = Kp
I = Kp / (Ti * s)
D = Kp * (Td * s) / (alpha * Td * s + 1)

# Controlador PID total
PID = P + I + D

# --- Simulação da resposta ao degrau unitário ---
t, y = ctrl.step_response(PID)  # Entrada degrau unitário é padrão

# --- Plotagem do gráfico ---
plt.figure(figsize=(10, 5))
plt.plot(t, y, 'b', linewidth=2)
plt.title('Resposta ao Degrau Unitário do Controlador PID com Filtro Derivativo')
plt.xlabel('Tempo (s)')
plt.ylabel('Saída')
plt.grid(True)
plt.tight_layout()
plt.show()
```
